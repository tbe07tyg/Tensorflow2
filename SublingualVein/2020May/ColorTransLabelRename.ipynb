```python
# raw label
raw_labels_root =  "E:\\dataset\\SublingualVein\\TIASRGB2020\\selected_masks\\raw"
print("raw_labels_root:", raw_labels_root)
```

```python
from glob import glob
raw_labels_paths =  sorted(glob(raw_labels_root+ "/*.png"))
print("raw_labels_paths", len(raw_labels_paths))
```

```python
import os
import shutil  # copy and rename file

output_folder_root = "E:\\dataset\\SublingualVein\\TIASRGB2020\\color_transfer_labels"
for path in raw_labels_paths:
    print("old_paht:", path)
    # copy and rename
    filename =  os.path.splitext(os.path.basename(path))[0]
    output_file_name = "color_tr_"+filename
    src = path
    dst = output_folder_root + "/" + output_file_name + ".png"
    print("new_path:", dst)
    shutil.copy(src, dst)
```
